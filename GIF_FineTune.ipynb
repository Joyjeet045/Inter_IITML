```python
system_message = """You are a specialized video question answering assistant. Your main tasks are to analyze video content, extract relevant information,
and answer questions based on the visual data present in the GIF. You can summarize key scenes, identify objects and actions, and
provide contextual information based on the content."""

def format_test_data(sample):
    return {"test_messages" : [
          {
              "role" : "system" ,
              "content" : [{"type" : "text" , "text" : system_message}]
          },
          {
              "role" : "user",
              "content" : [
                  {
                      "type" : "video",
                      "video" : sample["gif_url"]
                  },
                  {
                      "type" : "text",
                      "text" : sample['question']
                  }
              ]
          }
          ]}
```

```python
import pandas as pd
from datasets import Dataset

# count_csv = pd.read_csv('count.csv')
# actions_csv = pd.read_csv('action.csv')
# frameqa_csv = pd.read_csv('frameqa.csv')
transition_csv = pd.read_csv('transition.csv')
transition_csv.rename(columns = {'URL2':'gif_url'},inplace=True)

# count_csv.shape

# count_dataset = Dataset.from_pandas(count_csv)
# action_dataset = Dataset.from_pandas(actions_csv)
# frameqa_dataset = Dataset.from_pandas(frameqa_csv)
transition_dataset = Dataset.from_pandas(transition_csv)

# count_subset_train = count_dataset.select(range(500))
# action_subset_train = action_dataset.select(range(500))
# frameqa_subset_train = frameqa_dataset.select(range(500))
transition_subset_train = transition_dataset.select(range(500))

# count_subset_test = count_dataset.select(range(1000,1010))
# action_subset_test = action_dataset.select(range(500,510))
# frameqa_subset_test = frameqa_dataset.select(range(500,510))
transition_subset_test = transition_dataset.select(range(500,510))


# count_test_dt = [format_test_data(sample) for sample in count_subset_test]
# action_test_dt = [format_test_data(sample) for sample in action_subset_test]
transition_test_dt = [format_test_data(sample) for sample in transition_subset_test]
# frameqa_test_dt = [format_test_data(sample) for sample in frameqa_subset_test]
```

```python
from qwen_vl_utils import process_vision_info

def generate_answer(sample, model, processor):
    # Preparation for inference
    text = processor.apply_chat_template(
        sample["test_messages"], tokenize=False, add_generation_prompt=True
    )
    image_inputs, video_inputs = process_vision_info(sample["test_messages"])
    inputs = processor(
        text=[text],
        images=image_inputs,
        videos=video_inputs,
        padding=True,
        return_tensors="pt",
    )
    inputs = inputs.to(model.device)
    # Inference: Generation of the output
    generated_ids = model.generate(**inputs, max_new_tokens=60)
    generated_ids_trimmed = [
        out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]
    output_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )
    # print(output_text)
    return output_text

# let's generate the description
answer = generate_answer(transition_test_dt[0], model, processor)
print(answer)
```

```python
for i,sample in enumerate(transition_subset_test):
  print('Question : ' , sample['question'])
  print('Actual Answer : ' , sample['answer'])
  print('Answer : ' ,generate_answer(transition_test_dt[i], model, processor))

  print('GIF_URL : ' ,sample['gif_url'])
```
